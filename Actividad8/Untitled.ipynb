```python
#Eliminando datos innecesarios de "meteo"
none = []
for m in range(0, len(meteo.columns)):
    if(re.search("^Unnamed", meteo.columns[m]) != None):
        none.append(meteo.columns[m])
meteo = meteo.drop(none,1)
meteo = meteo.drop(0,0)
meteo.head()
```

```python
#Eliminando datos innecesarios de "suelo"
vacio0 = []
for m in range(0, len(suelo.columns)):
    if(re.search("^THETA", suelo.columns[m]) != None):
        vacio0.append(suelo.columns[m])
suelo = suelo.drop(vacio0,1)
suelo.head()
```
